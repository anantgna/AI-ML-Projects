```
# UDF Imputer

def imputer_age(cols):
  Age = cols[0]
  Pclass = cols[1]

  if pd.isnull(Age) :
    if Pclass == 1:
      return 37
    elif Pclass == 2 :
      return 29
    else :
      return 24
    
  else :
    return Age
```

```
# Apply function is used to apply Customized UDF 

train_data['Age'] = train_data[['Age', 'Pclass']].apply(imputer_age, axis = 1)
```
