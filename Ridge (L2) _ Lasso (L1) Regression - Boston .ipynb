```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
from sklearn.datasets import load_boston
```

```
boston = load_boston()
```

```
print(boston.DESCR)
```

```
boston_df = pd.DataFrame(data = boston.data, columns = boston.feature_names)
```

```
boston_df
```

```
boston_df['MEDV'] = boston.target
```

```
boston_df
```

```
boston_df.isnull().sum()
```

```
boston_df.shape
```

```
boston_df.info()
```

**Assumptions**

```
plt.figure(figsize=(12,8))
sns.distplot(boston_df['MEDV'], color = 'darkviolet')
plt.show()
```

**Correlation**

```
plt.figure(figsize = (15 ,10))
sns.heatmap(boston_df.corr(), annot = True, cmap = 'magma')
```

**X & Y Split**

```
X = boston_df.drop(['MEDV', 'RAD', 'CHAS'], axis = 1)
```

```
Y = boston_df['MEDV']
```

**Split Into Train & Test**

```
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 3)
```

### Lasso Regression

```
from sklearn.linear_model import Lasso, Ridge
```

```
lasso = Lasso(alpha = 0.0001, normalize = True, max_iter = 5000)
```

```
lasso.fit(X_train, Y_train)
```

```
lasso.score(X_test, Y_test)
```

```
# Lasso Coeffiecients - weights & biases of variable
lasso.coef_
```

```
lasso.intercept_
```

**Coefficient of Lasso (Feature Selection)**

```
sns.set_style('darkgrid')
plt.figure(figsize = (12,8))
plt.plot(range(len(X_test.columns)), lasso.coef_)
plt.xticks(range(len(X_test.columns)), X_test.columns, rotation = 90, fontsize = 12)
plt.show()
```

## Ridge Regression

```
ridge = Ridge(alpha = 0.001, normalize=True, max_iter=5000)
```

```
ridge.fit(X_train, Y_train)
```

```
ridge.score(X_test, Y_test)
```

**Metrics**

```
predicition = ridge.predict(X_test)
```

```
from sklearn.metrics import mean_squared_error, r2_score
```

```
mean_squared_error(Y_test, predicition)
```

```
r2_score(Y_test, predicition)
```
