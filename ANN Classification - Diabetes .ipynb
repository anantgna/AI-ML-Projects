# Importing Libraries & Dataset

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
diabetes_df = pd.read_csv('https://raw.githubusercontent.com/ammishra08/MachineLearning/master/Datasets/diabetes.csv')
```

```
diabetes_df
```

# Data Pre-processing

```
diabetes_df.info()
```

```
diabetes_df.isnull().sum()
```

```
diabetes_df.describe()
```

```
diabetes_df1 = diabetes_df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']].replace(0, np.nan)
```

```
diabetes_df1['Pregnancies'] = diabetes_df['Pregnancies']
```

```
diabetes_df1['Outcome'] = diabetes_df['Outcome']
```

```
diabetes_df1.isnull().sum()
```

```
diabetes_df1['Insulin'].unique()
```

```
diabetes_df1['SkinThickness'].unique()
```

```
mean = diabetes_df1['Insulin'].mean()
```

```
diabetes_df1['Insulin'] = diabetes_df1['Insulin'].replace(np.nan, mean)
```

```
diabetes_df1.isnull().sum()
```

```
from sklearn.impute import SimpleImputer
impute = SimpleImputer(strategy = 'median')
print(impute)
```

```
d_array = impute.fit_transform(diabetes_df1)
```

```
diabetes_df2 = pd.DataFrame(d_array, columns = diabetes_df1.columns)
```

```
diabetes_df2.isnull().sum()
```

```
diabetes_df2
```

```
sns.catplot(x= 'Outcome', kind = 'count', data = diabetes_df2)
```

```
diabetes_df2['Outcome'].value_counts()
```

```
from sklearn.utils import resample

df_0 = diabetes_df2[diabetes_df2['Outcome'] == 0.0]
df_1 = diabetes_df2[diabetes_df2['Outcome'] == 1.0]
```

```
df_1_upsample = resample(df_1, n_samples = 500, replace = True, random_state = 123)
```

```
diabetes_df3 = pd.concat([df_0, df_1_upsample])
```

```
sns.catplot(x= 'Outcome', kind = 'count', data = diabetes_df3.drop_duplicates())
```

```
sns.catplot(x= 'Outcome', kind = 'count', data = diabetes_df3)
```

```
diabetes_df3['Outcome'].value_counts()
```

# Data Visualization

```
plt.figure(figsize = (10,10))
sns.heatmap(diabetes_df3.corr(), annot = True, cmap = 'cool')
```

# Splitting Data

```
X = diabetes_df3.drop(['Outcome'], axis = 1) 
```

```
Y = diabetes_df3['Outcome']
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

# Building Model

```
# Dense Layer in NN
from keras.layers import Dense
# Feed Foreward NN
from keras.models import Sequential
# Model Optimizers
from keras.optimizers import Adam
```

```
def build_model ():
  model = Sequential()
  # Input Layer : num of neurons = (2)^n
  model.add(Dense(units = 64, activation='relu', input_shape = [len(X.keys())]))
  # Hidden Layer - I
  model.add(Dense(units = 128, activation='relu'))
  # Hidden Layer - II
  model.add(Dense(units = 256, activation='relu'))
  # Output Layer
  model.add(Dense(units = 1, activation='sigmoid'))  

  # Optimizers = Adam
  # Alpha = Learning Rate : sample size = small (0.001), sample size = large (0.01)
  optimizers = Adam(learning_rate= 0.0001)

  # For binary classification
  # if activation is sigmoid and o/p is binary : loss = 'binary_crossentropy'
  # if activation is softmax and o/p is binary : loss = 'categorical_crossentropy'
  model.compile(loss = 'binary_crossentropy', optimizer = optimizers, metrics = ['accuracy']) 
  
  return model 
```

```
model = build_model()
```

```
model.summary()
```

# Training Model

```
results = model.fit(X_train, Y_train, epochs = 400, batch_size = 25, validation_split = 0.20)
```

```
pd.DataFrame(results.history)
```

# Plotting Model

```
pd.DataFrame(results.history)[['accuracy', 'val_accuracy']].plot()
```

```
pd.DataFrame(results.history)[['loss', 'val_loss']].plot()
```

```
model.evaluate(X_test, Y_test)
```

# Prediction Using Model

```
predictions = model.predict(X_test)
```

```
yhat = np.round(predictions)
```

```
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test, yhat)
```

```
sns.heatmap(confusion_matrix(Y_test, yhat), annot = True, fmt='0.0f')
```

```
from sklearn.metrics import classification_report
print(classification_report(Y_test, yhat))
```
